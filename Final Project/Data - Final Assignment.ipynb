# Data

In order to address the problem, I collected data on the average house sale in London directly from [GOV.UK website](https://www.gov.uk/government/publications/uk-house-price-index-england-january-2020/uk-house-price-index-england-january-2020) and the Crime data from the [MET Police website](https://www.met.police.uk/sd/stats-and-data/met/crime-data-dashboard/).

Both were CSV file, so I just uploaded it to Jupyter Notebooks.

Once I cleaned up the data and created a usable data frame, I utilised  the [Foursquare's API](https://foursquare.com/) in order to obtain information about venues in each neighbourhood. 

For those who are not familiar with London and its neighbourhoods, I created a choropleth map utilising this [JSON](https://raw.githubusercontent.com/radoi90/housequest-data/master/london_boroughs.geojson) file and represented both, Crime and House Price on the map.
