# Introduction

London is among the oldest of the world’s great cities and one of the most cosmopolitan. 

Because of its position on the River Thames, it is the country’s economic, transportation, and cultural centre. London's docks, now home to trendy apartments, were the key to its prosperity as the world's greatest trading city in the 19th century, at the very height of the British Empire. 

London includes 31 neighbourhoods and two separate cities, ‘The City’ and the City of Westminster. 

London's many neighbourhoods radiate out of the centre and those who live there tend to define themselves as living north or south of the River Thames. In many ways these neighbourhoods are a series of villages, each with their own identity.

Buying a property in London can be the opportunity of a lifetime, but it is rarely simple or straightforward: besides the house price, other factors such as crime and what each neighbourhood can offer in terms of venues (restaurants, hotels, parks, …) also affect the decision-making process.

The aim of this project is to analyse each London’ neighbourhood and identify:
<ul>
<li>House affordability</li>
<li>Crime and safety</li>
<li>Top 10 venues that are most common in the neighbourhoods</li>
</ul>
